```python
# Baseline cadence in minion_1016_sqlite.db, downloadable from
# http://lsst.org/scientists/simulations/opsim/opsim-v335-benchmark-surveys
# MAF install instructions at:
# https://confluence.lsstcorp.org/display/SIM/Catalogs+and+MAF
import matplotlib.pyplot as plt
%matplotlib inline
import lsst.sims.maf.db as db
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.metricBundles as metricBundles
```

```python
# Set up the database connection
opsdb = db.OpsimDatabase('minion_1016_sqlite.db')
outDir = 'output'
resultsDb = db.ResultsDb(outDir=outDir)
```

```python
nside = 512 # Looks like the default being used
filters = ['u', 'g', 'r', 'i', 'z', 'y']
bundleList = []
slicer = slicers.HealpixSlicer(nside=nside, latCol='ditheredDec', lonCol='ditheredRA')
metric1 = metrics.Coaddm5Metric()
metric2 = metrics.CountMetric(col='expMJD', metricName='Number of Visits')
for filtername in filters:
    sql = 'filter = "%s"' % filtername
    bundleList.append(metricBundles.MetricBundle(metric1,slicer,sql))
    bundleList.append(metricBundles.MetricBundle(metric2,slicer,sql))
```

```python
bundleDict = metricBundles.makeBundlesDictFromList(bundleList)
bgroup = metricBundles.MetricBundleGroup(bundleDict, opsdb, outDir=outDir, resultsDb=resultsDb)
bgroup.runAll()
bgroup.plotAll(closefigs=False)
```
